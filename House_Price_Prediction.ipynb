Introduction

In this project, we aim to build a Machine Learning model using Linear Regression to predict house prices (in $1000s) based on the size of the house in square feet. The dataset used for this analysis contains information about house sizes and their corresponding prices. Linear Regression is a fundamental and widely-used algorithm in supervised learning, making it an ideal starting point for understanding the relationship between house size and price.

The goal is to create a model that can accurately predict house prices based on the input feature (size in square feet) and evaluate its performance using appropriate metrics. This project will also involve data preprocessing, visualization, and model evaluation to ensure the reliability and effectiveness of the predictions.

By the end of this notebook, we will have a trained Linear Regression model that can provide insights into how house size influences its price, which can be valuable for real estate analysis and decision-making.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

**numpy**: Used for numerical computations and data manipulation.

**matplotlib.pyplot**: Used for data visualization.

**sklearn.model_selection**: Used for splitting datasets into training and testing sets.

**sklearn.linear_model**: Used for implementing linear regression models.

```python
x = np.array([500, 750, 1000, 1250, 1500, 1750, 2000, 2250, 2500, 2750, 3000]).reshape(-1,1)
y = np.array([100, 150, 200, 250, 300, 350, 400, 450, 500, 550, 600])
```

**What it does:**

np.array([500, 750, 1000, ...]): Creates a NumPy array containing the house sizes in square feet.

.reshape(-1, 1): Reshapes the array into a 2D array with a single column. The -1 tells NumPy to automatically calculate the number of rows based on the data, and 1 specifies the number of columns.

**Purpose:**

X represents the independent variable (input feature) for the linear regression model. In this case, it is the size of the house in square feet.

The reshape(-1, 1) is necessary because scikit-learn expects the input feature (X) to be a 2D array (even if there's only one feature).

**Example:**

Before reshaping: [500, 750, 1000, ...] (1D array).

After reshaping: [[500], [750], [1000], ...] (2D array).

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

**What it does**: Splits the dataset (X and y) into random subsets for training and testing.

**Inputs:**

X: The independent variable (features), e.g., house sizes.

y: The dependent variable (target), e.g., house prices.

**test_size**: The proportion of the dataset to include in the test split. Here, test_size=0.2 means 20% of the data will be used for testing, and 80% will be used for training.

**random_state**: A seed value for the random number generator. Setting this ensures that the split is reproducible (i.e., you'll get the same split every time you run the code with the same seed).

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

**model = LinearRegression()**
What it does: Initializes an instance of the LinearRegression class.

Purpose: This creates a linear regression model object that will be used to learn the relationship between the independent variable (X_train) and the dependent variable (y_train).

Key Points:

The model starts with no knowledge of the data. It is an "empty" model at this stage.

The LinearRegression class implements ordinary least squares (OLS) regression, which minimizes the sum of squared errors between the predicted and actual values.

**model.fit(X_train, y_train)**
What it does: Trains the linear regression model on the training data (X_train and y_train).

Purpose: The fit method computes the optimal parameters (coefficients) for the linear regression model based on the training data.

How it works:

The model learns the relationship between the input feature (X_train, house sizes) and the target variable (y_train, house prices).

Mathematically, it finds the best-fit line (or hyperplane in higher dimensions) that minimizes the difference between the predicted and actual values.

Key Points:

After training, the model will have learned the coefficients (slope and intercept) of the linear equation:

y
=
intercept
+
slope
⋅
X
y=intercept+slope⋅X
These coefficients can be accessed using model.intercept_ and model.coef_.

```python
y_predict = model.predict(x_test)

print(f"predicted prices: {y_predict}")
print(f"Actual prices: {y_test}")
```

```python
# Plot training data
plt.scatter(x_train, y_train, color='blue', label="Training Data")
plt.scatter(x_test, y_test, color='red', label="Test Data")
plt.plot(x, model.predict(x), color='green', linestyle='dashed', label="Regression Line")

plt.xlabel("House Size (sq ft)")
plt.ylabel("Price ($1000s)")
plt.legend()
plt.title("House Prices vs Size")
plt.show()
```
